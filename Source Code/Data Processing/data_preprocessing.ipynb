## Top 3 important factors affecting the Employee Performance


```python
# heatmap
plt.figure(figsize=(25, 25))
sns.heatmap(df.corr(), annot=True, fmt=".1f", cmap="Blues")
sns.set(font_scale=2)
plt.show()
```

### Visualizing Variable Relationships with a Heatmap

- A heatmap is an effective tool to visualize correlations between variables. By mapping data values onto a color gradient, it provides an intuitive way to interpret relationships and trends within a dataset.

- Representing data in a two-dimensional matrix or table with color changes to depict value magnitudes.

- Highlighting similarities and differences using color gradients and clustering techniques.

- The color gradient on the heatmap can be used to represent the strength of the correlation (positive or negative) between variables. Darker shades indicate stronger correlations, while lighter shades indicate weaker correlations.
- When clustering is applied, the grouped data is displayed on the heatmap, enhancing the understanding of relationships among variables.


```python
# displays the correlated data
corr_matrix = df.corr()
corr_matrix["PerformanceRating"].sort_values(ascending=False)
```

---

From the correlation analysis, we can observe both positively correlated and negatively correlated factors influencing employee performance. Among these, the top three factors that significantly impact employee performance are:

- _Employee Environment Satisfaction_
- _Employee Last Salary Hike Percentage_
- _Employee Work-Life Balance_

These factors highlight the importance of fostering a positive work environment, offering fair salary increments, and ensuring a healthy work-life balance to enhance employee performance and satisfaction.
