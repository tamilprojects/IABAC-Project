```python
# select the columns with objects that are unique
df.select_dtypes(include=["object"]).nunique()
```

```python
# Exclude 'EmpNumber' and get value counts for the remaining categorical columns
categorical_columns = df.select_dtypes(include=["object"]).columns

# Filter out 'EmpNumber'
categorical_columns = [col for col in categorical_columns if col != "EmpNumber"]

# Iterate and print value counts for remaining columns
for col in categorical_columns:
    print(f"Value counts for column '{col}':")
    print(df[col].value_counts())
    print("-" * 40)
```
