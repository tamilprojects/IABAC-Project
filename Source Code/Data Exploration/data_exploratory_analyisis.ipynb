> # Domain Analysis

Understanding the significance of each feature and evaluating its impact on the dependent feature, **PerformanceRating**, is critical to meaningful insights. Below is an explanation of each feature in the dataset:

1. **EmpNumber**:
   A unique identifier assigned to each employee within the organization, ensuring distinct records for analysis.

1. **Age**:
   Represents the age of the employee in years. Age may correlate with experience, adaptability, and performance.

   Denotes the gender of the employee, categorized as **Male** or **Female**. Understanding gender trends can reveal diversity insights.

1. **EducationBackground**:
   Indicates the highest academic qualification achieved by the employee, such as a high school diploma, degree, or postgraduate degree.

1. **MaritalStatus**:  
   Refers to the civil status of the employee, such as **Single**, **Married**, or **Divorced**. It may have indirect impacts on work-life balance or relocation preferences.

1. **EmpDepartment**:
   Specifies the department to which the employee belongs, such as IT, HR, or Finance, potentially influencing role-specific performance metrics.

1. **EmpJobRole**:
   Represents the key responsibilities and expectations associated with the employee's job profile or position.

1. **BusinessTravelFrequency**:
   Indicates the frequency of travel undertaken by the employee for business purposes, which may affect stress levels and performance.

1. **DistanceFromHome**:
   The distance between the employee’s residence and the workplace. Longer commutes may impact work-life balance and satisfaction.

1. **EmpEducationLevel**:
   Reflects the level of academic achievement, such as **Diploma**, **Degree**, or **Ph.D.**, which may influence skills and performance.

1. **EmpEnvironmentSatisfaction**:
   Measures how content employees are with their work environment, including job conditions and organizational culture.

1. **EmpHourlyRate**:
   The hourly wage of the employee, indicating their compensation for each hour worked.

1. **EmpJobInvolvement**
   Represents the degree of psychological identification with the job, affecting motivation and productivity.

1. **EmpJobLevel**:
   Denotes the classification or grade of the job, which determines the level of responsibility and expectations.

1. **EmpJobSatisfaction**:
   Assesses how satisfied employees are with their job roles, impacting overall engagement and performance.

1. **NumCompaniesWorked**:
   Indicates the number of companies the employee has previously worked for, which may reflect experience and adaptability.

1. **OverTime**:
   Whether the employee works overtime. Excessive overtime could lead to stress, while reasonable levels may indicate commitment.

1. **EmpLastSalaryHikePercent**:
   The percentage of the last salary increment, potentially reflecting employee satisfaction and loyalty.

1. **EmpRelationshipSatisfaction**:
   Measures satisfaction with workplace relationships, impacting collaboration and morale.

1. **TotalWorkExperienceInYears**:
   Total years of professional experience, which often correlates with expertise and performance.

1. **TrainingTimesLastYear**:
   The number of training sessions attended by the employee in the previous year, contributing to skill enhancement.

1. **EmpWorkLifeBalance**:
   Indicates the balance between professional and personal life. A healthy work-life balance promotes motivation and reduces stress.

1. **ExperienceYearsAtThisCompany**:
   Total years of experience the employee has within the organization, influencing loyalty and institutional knowledge.

1. **ExperienceYearsInCurrentRole**:
   Total years of experience in the employee's current role, reflecting role-specific expertise.

1. **YearsSinceLastPromotion**:
   The number of years since the employee’s last promotion, potentially impacting morale and aspirations.

1. **YearsWithCurrManager**:
   The duration of the employee’s working relationship with their current manager, affecting trust and collaboration.

1. **Attrition**:
   Indicates whether the employee has voluntarily left the organization. Factors influencing attrition include satisfaction, career growth, and compensation.

1. **PerformanceRating**:
   The dependent feature representing the overall performance score of an employee within the company. This serves as the target variable for analysis.

# Using Subplots for the multiple features


```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of selected features for univariate analysis
count = data[
    [
        "Gender",
        "EducationBackground",
        "MaritalStatus",
        "BusinessTravelFrequency",
        "DistanceFromHome",
        "EmpEducationLevel",
        "EmpEnvironmentSatisfaction",
        "EmpJobInvolvement",
        "EmpJobLevel",
        "EmpJobSatisfaction",
        "NumCompaniesWorked",
        "OverTime",
    ]
]

# Set the figure size
plt.figure(figsize=(20, 25))

# Counter for subplots
plotno = 1

# Iterate through each column in the 'count' DataFrame
for column in count:
    if plotno <= 12:  # Limit the number of plots to 12
        plt.subplot(4, 3, plotno)  # Positioning the subplot in a 4x3 grid

        # Create a countplot for each categorical feature without 'hue'
        sns.countplot(x=count[column], palette="rocket")

        # Set x-axis label with appropriate font size
        plt.xlabel(column, fontsize=12)

        # Set the title for each subplot to identify the feature
        plt.title(f"Distribution of {column}", fontsize=20)

    plotno += 1  # Increment counter for next subplot

# Adjust the layout to ensure plots don't overlap
plt.tight_layout()

# Display the plots
plt.show()
```

#### CREATE A NEW DATAFRAME FOR CATEGORICAL FEATURES.


```python
categorical = []  # Create empty list to store the feature
for column in data.columns:
    if data[column].dtype == object:
        categorical.append(column)
        print(f"{column} : {data[column].unique()}")
        print("====================================")
```

```python
categorical  # All categorical features
```

```python
# Make a list of categorical feature and use sub plot to plot multiple feature
categorical_col = data[
    [
        "Gender",
        "EducationBackground",
        "MaritalStatus",
        "BusinessTravelFrequency",
        "OverTime",
        "Attrition",
    ]
]
plt.figure(figsize=(20, 22))  # defining canvas size
plotno = 1  # counter

for column in categorical_col:  # iteration of columns / acessing the columns from count
    if plotno <= 7:  # set the limit
        plt.subplot(
            3, 2, plotno
        )  # # plotting 6 graphs (3-rows,2-columns) ,plotnumber is for count
        sns.countplot(
            x=categorical_col[column], palette="rocket"
        )  # Plotting count plots because the feature is categorical
        plt.xlabel(column, fontsize=20)  # assigning name to x-axis and font size is 20
        plt.ylabel("PerformanceRating")
    plotno += 1  # counter increment
plt.tight_layout()
plt.show()  # used to hide the storage loction
```

**13.Relationship between employee department and performance rating**


```python
# Calculate the order of EmpDepartment based on counts
order = data["EmpDepartment"].value_counts().sort_values(ascending=False).index

plt.figure(figsize=(20, 10))
sns.countplot(x="EmpDepartment", data=data, palette="rocket", order=order)
plt.xlabel("EmpDepartment", fontsize=20)
plt.ylabel("PerformanceRating", fontsize=15)
plt.tight_layout()
plt.show()
```

**14.Relationship between employee job role and performance rating**
